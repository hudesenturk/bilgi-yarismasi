```python
sorular = ["Kızıl Gezegen genellikle hangisi için yapılan bir benzetme olur?",
           "Atlas Dağları'nı gezmekte olan biri hangi ülkede olabilir?",
           "Kimlere karıncaezmez denir?",
           "Osmanlı tarihinde, hangi addaki padişahların toplam hükümdarlık süresi yaklaşık yüz yılla en uzundur?",
           "Guinness Dünya Rekorlarına göre Sri Lanka'da bulunan ve insan eliyle dikilmiş bilinen en eski ağaç olan Sri Maha Bodhiya adlı ağaç hangi tarihte dikilmiştir?"]

soru_siklari = [["A) Kripton", "B) Cyberton", "C) Mars", "D) Vulcan"],
                ["A) Arjantin", "B) Fas", "C) Arnavutluk", "D) Japonya"],
                ["A) Çok esprili kişilere", "B) Çok merhametli kişilere", "C) Çok gösterişli kişilere", "D) Çok çalışkan kişilere"],
                ["A) Mehmed", "B) Selim", "C) Mahmud", "D) Murad"],
                ["A) MÖ 1288", "B) MÖ 288", "C) MS 288", "D) MS 1288"]]

dogru_cevaplar = ["c", "b", "b", "a", "b"]

print("BİLGİ YARIŞMASINA HOŞ GELDİNİZ")

puan = 0

for i in range(5):
    print(f"Soru {i + 1}: {sorular[i]}")
    for sik in soru_siklari[i]:
        print(sik)
    cevap = input("Cevabınız nedir: ").lower()
    if cevap == dogru_cevaplar[i]:
        puan += 2000
        print(f"Tebrikler doğru cevap verdiniz, {puan} TL kazandınız!")
    else:
        print(f"Yanlış cevap verdiniz, doğru cevap {dogru_cevaplar[i]} şıkkı.")
        break

if puan == 10000:
    print("Tebrikler! 10.000 TL kazandınız.")
else:
    print(f"Yarışmadan {puan} TL ile ayrıldınız.")
```
